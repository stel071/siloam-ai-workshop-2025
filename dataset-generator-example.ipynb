# MCU Data

```python
import requests

url = "https://mcu-api.siloamhospitals.com/mcu-v2/api/v4/package-search-list?hospital=siloam-hospitals-lippo-village&page=0&count=500&application=mcu&apiVersion=v4&serviceType=walkin"
response = requests.get(url)
response_json = response.json()['data']
```

```python
display(response_json['content'][0])
```

```python
mcu_packages = [{'id': item['id'], 'name': item['name_en'], 'description': item['description_en'], 'price': item['starting_price']} for item in response_json['content']]
```

```python
# write data to package.json
import json
with open("mcu.json", "w") as f:
    json.dump(mcu_packages, f, indent=4)
```

```python
# move data to sqlite file
import sqlite3
conn = sqlite3.connect('mcu.db')
c = conn.cursor()
c.execute('''
    CREATE TABLE IF NOT EXISTS patients (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER NOT NULL
    )
''')
c.execute('''
    CREATE TABLE IF NOT EXISTS package (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        description TEXT NOT NULL,
        price INTEGER NOT NULL
    )
''')
c.execute('''
    CREATE TABLE IF NOT EXISTS transactions (
        id INTEGER PRIMARY KEY,
        patient_id INTEGER NOT NULL,
        package_id INTEGER NOT NULL,
        date TEXT NOT NULL,
        status TEXT NOT NULL,
        FOREIGN KEY (patient_id) REFERENCES patients (id),
        FOREIGN KEY (package_id) REFERENCES package (id)
    )
''')
c.executemany('INSERT INTO patients (id, name, age) VALUES (?, ?, ?)', [(p['id'], p['name'], p['age']) for p in patients])
c.executemany('INSERT INTO package (id, name, description, price) VALUES (?, ?, ?, ?)', [(p['id'], p['name'], p['description'], p['price']) for p in package])
c.executemany('INSERT INTO transactions (id, patient_id, package_id, date, status) VALUES (?, ?, ?, ?, ?)', [(t['id'], t['patient_id'], t['package_id'], t['date'], t['status']) for t in transactions])
conn.commit()
conn.close()    
```
